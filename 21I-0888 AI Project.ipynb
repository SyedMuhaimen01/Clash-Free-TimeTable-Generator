```python
import random
from random import randint

# Initializing the maximum values of chromosome components
courses=5
sections=16
teachers=20
rooms=20
classType=1
max_students=120
timeSlots=6
roomSizeMax=120
days=5

# room with capcaity of 120 students
halls = random.sample(range(1, rooms + 1), rooms // 3) 

# calculating total number of bits needed to store each one of these values
courses_bits = int(courses.bit_length())
sections_bits = int(sections.bit_length())
teachers_bits = int(teachers.bit_length())
rooms_bits = int(rooms.bit_length())
classType_bits = int(classType.bit_length())
max_students_bits = int(max_students.bit_length())
timeSlots_bits = int(timeSlots.bit_length())
roomSizeMax_bits = int(roomSizeMax.bit_length())
days_bits = int(days.bit_length())

# creating bits to store the original data
courses_binary = '0' * courses_bits
sections_binary = '0' * sections_bits
teachers_binary = '0' * teachers_bits
rooms_binary = '0' * rooms_bits
classType_binary = '0' * classType_bits
max_students_binary = '0' * max_students_bits
timeSlots_binary = '0' * timeSlots_bits
roomSizeMax_binary = '0' * roomSizeMax_bits
days_binary = '0' * days_bits

# converting binaries from string to int
courses_binary_int = int(courses_binary, 2)
sections_binary_int = int(sections_binary, 2)
teachers_binary_int = int(teachers_binary, 2)
rooms_binary_int = int(rooms_binary, 2)
classType_binary_int = int(classType_binary, 2)
max_students_binary_int = int(max_students_binary, 2)
timeSlots_binary_int = int(timeSlots_binary, 2)
roomSizeMax_binary_int = int(roomSizeMax_binary, 2)
days_binary_int = int(days_binary, 2)




# allocate teacher
allocateTeacher= {i: [] for i in range(1, teachers + 1)}

# assigning teacher
assignTeacher = {i: {j: 0 for j in range(1, courses + 1)} for i in range(1, teachers + 1)}

# initial population of random bitstring
pop = sections*courses
chromosome=[]
fitness = {i: 0 for i in range(1, pop + 1)}

#function that generates the intial population of chromosome
def Initialize_population():
    
    #accessing the chromosome population
    global chromosome
    for section in range(sections):
        # number of assigned courses to a teacher
        teacher_course_count = {i: 0 for i in range(1, teachers + 1)}  
        for course in range(courses):
            
            # hall or normal class
            classType = randint(0, 1) 
            
            sectionNumber = section + 1
            teacher = randint(1, teachers)
            
            # class strength
            class_strength = randint(45, max_students)
            
            # assigning day
            day1 = randint(1, days)
            day2 = randint(1, days)
            
            # assiging time slot
            timeSlot1 = randint(1, timeSlots)
            timeSlot2 = timeSlot1
            
            # assigning room
            room_size1 = 0
            while room_size1 < class_strength:
                room1 = randint(1, rooms)
                if room1 in halls:
                    room_size1 = 120
                else:
                    room_size1 = 60

            room2 = randint(1, rooms)
            room_size2 = 0
            while room_size2 < class_strength:
                room2 = randint(1, rooms)
                if room2 in halls:
                    room_size2 = 120
                else:
                    room_size2 = 60


            # Check if the teacher can teach more courses and assign the course
            if teacher_course_count[teacher] < 3:
                teacher_course_count[teacher] += 1
                teacherIndex = teacher
            else:
                print("Teacher's workload full. Assigning course to another teacher")
                # Choose another teacher randomly
                while True:
                    new_teacher = randint(1, teachers)
                    if teacher_course_count[new_teacher] < 3:
                        teacher = new_teacher
                        teacherIndex = teacher
                        teacher_course_count[teacher] += 1
                        break

            # Append course number, section number, and day to teacherAllocation
            allocateTeacher[teacherIndex].append((course + 1, sectionNumber, day1))
            allocateTeacher[teacherIndex].append((course + 1, sectionNumber, day2))

            # Increment teacherAssignment for the course
            assignTeacher[teacherIndex][course + 1] += 1

            # craeting binary representations
            # course
            course = bin(course+1)[2:]
            i = courses_bits
            k=0
            j=len(course)
            while i>0:
                i -= 1
                if j >= len(course):
                    if course[k] == "1" and courses_binary[k] == "1":
                        course = course[:k] + "1" + course[k+1:]            
                else:
                    course = course[:k] + "0" + course[k+1:]
    
                while i==0 and len(course) < courses_bits:  
                    course = "0" + course
                
            # classType
            classType = bin(classType)[2:]
            i = classType_bits
            k=0
            j=len(classType)
            while i>0:
                i -= 1
                if j >= len(classType):
                    if classType[k] == "1" and classType_binary[k] == "1":
                        classType = classType[:k] + "1" + classType[k+1:]            
                else:
                    classType = classType[:k] + "0" + classType[k+1:]
    
                while i==0 and len(classType) < classType_bits:  
                    classType = "0" + classType
            # sectionNumber
            sectionNumber = bin(sectionNumber)[2:]
            i = sections_bits
            k=0
            j=len(sectionNumber)
            while i>0:
                i -= 1
                if j >= len(sectionNumber):
                    if sectionNumber[k] == "1" and sections_binary[k] == "1":
                        sectionNumber = sectionNumber[:k] + "1" + sectionNumber[k+1:]            
                else:
                    sectionNumber = sectionNumber[:k] + "0" + sectionNumber[k+1:]
    
                while i==0 and len(sectionNumber) < sections_bits:  
                    sectionNumber = "0" + sectionNumber
                
            # class strength
            class_strength = bin(class_strength)[2:]
            i = max_students_bits
            k=0
            j=len(class_strength)
            while i>0:
                i -= 1
                if j >= len(class_strength):
                    if class_strength[k] == "1" and sections_binary[k] == "1":
                        class_strength = class_strength[:k] + "1" + class_strength[k+1:]            
                else:
                    class_strength = class_strength[:k] + "0" + class_strength[k+1:]
    
                while i==0 and len(class_strength) < max_students_bits:  
                    class_strength = "0" + class_strength
            
            # teacher
            teacher = bin(teacher)[2:]
            i = teachers_bits
            k=0
            j=len(teacher)
            while i>0:
                i -= 1
                if j >= len(teacher):
                    if teacher[k] == "1" and teachers_binary[k] == "1":
                        teacher = teacher[:k] + "1" + teacher[k+1:]            
                else:
                    teacher = teacher[:k] + "0" + teacher[k+1:]
    
                while i==0 and len(teacher) < teachers_bits:  
                    teacher = "0" + teacher
                
            # day1
            day1 = bin(day1)[2:]
            i = days_bits
            k=0
            j=len(day1)
            while i>0:
                i -= 1
                if j >= len(day1):
                    if day1[k] == "1" and days_binary[k] == "1":
                        day1 = day1[:k] + "1" + day1[k+1:]            
                else:
                    day1 = day1[:k] + "0" + day1[k+1:]
    
                while i==0 and len(day1) < days_bits:  
                    day1 = "0" + day1  
                
            # timeSlot1
            timeSlot1 = bin(timeSlot1)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot1)
            while i>0:
                i -= 1
                if j >= len(timeSlot1):
                    if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
                else:
                    timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]
    
                while i==0 and len(timeSlot1) < timeSlots_bits:  
                    timeSlot1 = "0" + timeSlot1 
                
            # room1
            room1 = bin(room1)[2:]
            i = rooms_bits
            k=0
            j=len(room1)
            while i>0:
                i -= 1
                if j >= len(room1):
                    if room1[k] == "1" and rooms_binary[k] == "1":
                        room1 = room1[:k] + "1" + room1[k+1:]            
                else:
                    room1 = room1[:k] + "0" + room1[k+1:]
    
                while i==0 and len(room1) < rooms_bits:  
                    room1 = "0" + room1     
            
            # room_size1
            room_size1 = bin(room_size1)[2:]
            i = roomSizeMax_bits
            k=0
            j=len(room_size1)
            while i>0:
                i -= 1
                if j >= len(room_size1):
                    if room_size1[k] == "1" and roomSizeMax_binary[k] == "1":
                        room_size1 = room_size1[:k] + "1" + room_size1[k+1:]            
                else:
                    room_size1 = room_size1[:k] + "0" + room_size1[k+1:]
    
                while i==0 and len(room_size1) < roomSizeMax_bits:  
                    room_size1 = "0" + room_size1   
                
            # day2
            day2 = bin(day2)[2:]
            i = days_bits
            k=0
            j=len(day2)
            while i>0:
                i -= 1
                if j >= len(day2):
                    if day2[k] == "1" and days_binary[k] == "1":
                        day2 = day2[:k] + "1" + day2[k+1:]            
                else:
                    day2 = day2[:k] + "0" + day2[k+1:]
    
                while i==0 and len(day2) < days_bits:  
                    day2 = "0" + day2 
            
            # timeSlot2
            timeSlot2 = bin(timeSlot2)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]
    
                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2 
                
            # room2
            room2 = bin(room2)[2:]
            i = rooms_bits
            k=0
            j=len(room2)
            while i>0:
                i -= 1
                if j >= len(room2):
                    if room2[k] == "1" and rooms_binary[k] == "1":
                        room2 = room2[:k] + "1" + room2[k+1:]            
                else:
                    room2 = room2[:k] + "0" + room2[k+1:]
    
                while i==0 and len(room2) < rooms_bits:  
                    room2 = "0" + room2   
            # room_size2
            room_size2 = bin(room_size2)[2:]
            i = roomSizeMax_bits
            k=0
            j=len(room_size2)
            while i>0:
                i -= 1
                if j >= len(room_size2):
                    if room_size2[k] == "1" and roomSizeMax_binary[k] == "1":
                        room_size2 = room_size1[:k] + "1" + room_size2[k+1:]            
                else:
                    room_size2 = room_size2[:k] + "0" + room_size2[k+1:]
    
                while i==0 and len(room_size2) < roomSizeMax_bits:  
                    room_size2 = "0" + room_size2 
               
            #create chromosome
            chromosome.append(course + classType + sectionNumber + class_strength + teacher + day1 + timeSlot1 + room1 + room_size1 + day2 + timeSlot2 + room2 + room_size2)
            
            
# calculate fitness
def calculateFitness(source_chromosome, population):
    clashes = 0
    
    for other_chromosome in population:
        
        
        # Extract individual genes of source chromosome
        source_chromosome_course = source_chromosome[: courses_bits]
        source_chromosome_classType = source_chromosome[courses_bits : (courses_bits + classType_bits)]
        source_chromosome_section = source_chromosome[(courses_bits + classType_bits) : (courses_bits + classType_bits + sections_bits)]
        source_chromosome_class_strength = source_chromosome[(courses_bits + classType_bits + sections_bits) : (courses_bits + classType_bits + sections_bits + max_students_bits)]
        source_chromosome_teacher = source_chromosome[(courses_bits + classType_bits + sections_bits+max_students_bits) : (courses_bits + classType_bits + sections_bits+max_students_bits + teachers_bits)]
        source_chromosome_day1 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits)]
        source_chromosome_timeSlot1 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits)]
        source_chromosome_room1 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits)]
        source_chromosome_room_size1 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)]
        source_chromosome_day2 = source_chromosome[(courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits) : (courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits)]
        source_chromosome_timeSlot2 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits)]
        source_chromosome_room2 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits)]
        source_chromosome_room_size2 = source_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)]
        
        # Extract individual genes of other chromosome
        other_chromosome_course = other_chromosome[: courses_bits]
        other_chromosome_classType = other_chromosome[courses_bits : (courses_bits + classType_bits)]
        other_chromosome_section = other_chromosome[(courses_bits + classType_bits) : (courses_bits + classType_bits + sections_bits)]
        other_chromosome_class_strength = other_chromosome[(courses_bits + classType_bits + sections_bits) : (courses_bits + classType_bits + sections_bits + max_students_bits)]
        other_chromosome_teacher = other_chromosome[(courses_bits + classType_bits + sections_bits+max_students_bits) : (courses_bits + classType_bits + sections_bits+max_students_bits + teachers_bits)]
        other_chromosome_day1 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits)]
        other_chromosome_timeSlot1 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits)]
        other_chromosome_room1 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits)]
        other_chromosome_room_size1 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)]
        other_chromosome_day2 = other_chromosome[(courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits) : (courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits)]
        other_chromosome_timeSlot2 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits)]
        other_chromosome_room2 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits)]
        other_chromosome_room_size2 = other_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)]
        

        #if same chromosome then skip
        if(source_chromosome_course==other_chromosome_course and source_chromosome_section==other_chromosome_section):
            continue
        # Calculate clashes

        # Classes on the same day and adjacent days are not allowed
        if int(source_chromosome_day1) == int(source_chromosome_day2) or int(source_chromosome_day2) == int(source_chromosome_day1) + 1:
            clashes += 1
        if (source_chromosome_class_strength > source_chromosome_room_size1):
            clashes += 1
        if (source_chromosome_class_strength > source_chromosome_room_size2):
            clashes += 1
        if (source_chromosome_section == other_chromosome_section and 
            source_chromosome_timeSlot1 == other_chromosome_timeSlot1 and
            source_chromosome_day1 == other_chromosome_day1):
            clashes += 1
        if (source_chromosome_section == other_chromosome_section and 
            source_chromosome_timeSlot2 == other_chromosome_timeSlot2 and
            source_chromosome_day2 == other_chromosome_day2):
            clashes += 1

        # Clash in teacher's schedule
        if (source_chromosome_day1 == other_chromosome_day1 and 
            source_chromosome_timeSlot1 == other_chromosome_timeSlot1 and 
            source_chromosome_teacher == other_chromosome_teacher):
            clashes += 1

        if (source_chromosome_day2 == other_chromosome_day2 and 
            source_chromosome_timeSlot2 == other_chromosome_timeSlot2 and 
            source_chromosome_teacher == other_chromosome_teacher):
            clashes += 1

        # Clash on the same day in the same room in the same time slot
        if (source_chromosome_day1 == other_chromosome_day1 and 
            source_chromosome_timeSlot1 == other_chromosome_timeSlot1 and 
            source_chromosome_room1 == other_chromosome_room1):
            clashes += 1

        if (source_chromosome_day2 == other_chromosome_day2 and 
            source_chromosome_timeSlot2 == other_chromosome_timeSlot2 and 
            source_chromosome_room2 == other_chromosome_room2):
            clashes += 1

        #soft constraints
        if(source_chromosome_classType==1 and source_chromosome_timeSlot1<4):
            clashes+=1
        if(source_chromosome_classType==0 and source_chromosome_timeSlot1>4):
            clashes+=1
        if(source_chromosome_classType==0 and source_chromosome_timeSlot2>4):
            clashes+=1
        
        if(source_chromosome_room1!=source_chromosome_room2):
            clashes+=1
    return -clashes  # Fitness function should be the inverse of the conflicts/clashes


    
def tournament_selection(fitness, tournament_size):
    
    selected = []
    # Select 2 worst chromosomes
    while len(selected) < 2:  
        tournament_individuals = random.sample(list(fitness.keys()), tournament_size)
        worst_individuals = sorted(tournament_individuals, key=lambda x: fitness[x])[:2]
        selected.extend(worst_individuals)
    return selected



def perform_crossover(chromosome1, chromosome2):
    point1 = randint((courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits),
                     ((courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits)-1))
    point2 = randint((courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits),
                     ((courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits)-1))

    
   
    # Performing crossover
    child1 = chromosome1[:point1] + chromosome2[point1:point2] + chromosome1[point2:]
    child2 = chromosome2[:point1] + chromosome1[point1:point2] + chromosome2[point2:]
    
    day1 = child1[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits)]
    room1 = child1[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits)]
    room_size1=0
    if room1 in halls:
        room_size1 = 120
    else:
        room_size1 = 60
        
    # room_size1
    room_size1 = bin(room_size1)[2:]
    i = roomSizeMax_bits
    k=0
    j=len(room_size1)
    while i>0:
        i -= 1
        if j >= len(room_size1):
            if room_size1[k] == "1" and roomSizeMax_binary[k] == "1":
                room_size1 = room_size1[:k] + "1" + room_size1[k+1:]            
        else:
            room_size1 = room_size1[:k] + "0" + room_size1[k+1:]
    
        while i==0 and len(room_size1) < roomSizeMax_bits:  
            room_size1 = "0" + room_size1
                
    room2 = child1[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits) : ]
    #convert to binary
    room_size2=0
    if room2 in halls:
        room_size2 = 120
    else:
        room_size2 = 60
        
    # room_size2
    room_size2 = bin(room_size2)[2:]
    i = roomSizeMax_bits
    k=0
    j=len(room_size2)
    while i>0:
        i -= 1
        if j >= len(room_size2):
            if room_size2[k] == "1" and roomSizeMax_binary[k] == "1":
                room_size2 = room_size1[:k] + "1" + room_size2[k+1:]            
        else:
            room_size2 = room_size2[:k] + "0" + room_size2[k+1:]
    
        while i==0 and len(room_size2) < roomSizeMax_bits:  
            room_size2 = "0" + room_size2
    
    timeSlot1 = child1[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits)]
    # Check and adjust timeSlot if necessary for child1
    if child1[courses_bits] == '1':
        if int(timeSlot1, 2) == 6:
            if int(timeSlot1, 2) == 6:
            
                # timeSlot1
                timeSlot1 = bin(5)[2:]
                i = timeSlots_bits
                k=0
                j=len(timeSlot1)
                while i>0:
                    i -= 1
                    if j >= len(timeSlot1):
                        if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                            timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
                    else:
                        timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]

                    while i==0 and len(timeSlot1) < timeSlots_bits:  
                        timeSlot1 = "0" + timeSlot1
                # timeSlot2
                timeSlot2 = bin(int(timeSlot1, 2) + 1)[2:]
                i = timeSlots_bits
                k=0
                j=len(timeSlot2)
                while i>0:
                    i -= 1
                    if j >= len(timeSlot2):
                        if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                            timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                    else:
                        timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                    while i==0 and len(timeSlot2) < timeSlots_bits:  
                        timeSlot2 = "0" + timeSlot2 
            child1 = child1[:(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits)] + timeSlot1 + child1[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)] + day1 + timeSlot2 + room1 +room_size1
        else:
            # timeSlot2
            timeSlot2 = bin(int(timeSlot1, 2) + 1)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2
                    
            child1 = child1[:(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits)] + timeSlot1 + child1[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)] + day1 + timeSlot2 + room1 +room_size1

    
    # Check and adjust timeSlot if necessary for child2
    if child2[courses_bits] == '1':
        if int(timeSlot1, 2) == 6:
            # timeSlot1
            timeSlot1 = bin(5)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot1)
            while i>0:
                i -= 1
                if j >= len(timeSlot1):
                    if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
                else:
                    timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]

                while i==0 and len(timeSlot1) < timeSlots_bits:  
                    timeSlot1 = "0" + timeSlot1
            # timeSlot2
            timeSlot2 = bin(int(timeSlot1, 2) + 1)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2
            child2 = child2[:(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits)] + timeSlot1 + child2[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)] + day1 + timeSlot2 + room1 + room_size1

        else:
            # timeSlot2
            timeSlot2 = bin(int(timeSlot1, 2) + 1)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2
            child2 = child2[:(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits)] + timeSlot1 + child2[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)] + day1 + timeSlot2 + room1 + room_size1

        # checking boundary values
        # for time slots
        if int(timeSlot1, 2)> timeSlots:
        # timeSlot1
            timeSlot1 = bin(5)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot1)
            while i>0:
                i -= 1
                if j >= len(timeSlot1):
                    if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
                else:
                    timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]

                while i==0 and len(timeSlot1) < timeSlots_bits:  
                    timeSlot1 = "0" + timeSlot1
        
        if int(timeSlot2, 2)> timeSlots:
        # timeSlot2
            timeSlot2 = bin(5)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2
         # for rooms          
    if int(room1, 2) > rooms :
        # room1
        room1 = bin(rooms)[2:]
        i = rooms_bits
        k=0
        j=len(room1)
        while i>0:
            i -= 1
            if j >= len(room1):
                if room1[k] == "1" and rooms_binary[k] == "1":
                    room1 = room1[:k] + "1" + room1[k+1:]            
            else:
                room1 = room1[:k] + "0" + room1[k+1:]

            while i==0 and len(room1) < rooms_bits:  
                room1 = "0" + room1     
            
    if int(room2, 2) > rooms :
        # room2
        room2 = bin(rooms)[2:]
        i = rooms_bits
        k=0
        j=len(room2)
        while i>0:
            i -= 1
            if j >= len(room2):
                if room2[k] == "1" and rooms_binary[k] == "1":
                    room2 = room2[:k] + "1" + room2[k+1:]            
            else:
                room2 = room2[:k] + "0" + room2[k+1:]

            while i==0 and len(room2) < rooms_bits:  
                room2 = "0" + room2  
   
        
    
    return child1, child2

   

def perform_mutation(test_chromosome):
    
    
    test_chromosome_course = test_chromosome[: courses_bits]
    test_chromosome_classType = test_chromosome[courses_bits : (courses_bits + classType_bits)]
    test_chromosome_section = test_chromosome[(courses_bits + classType_bits) : (courses_bits + classType_bits + sections_bits)]
    test_chromosome_class_strength = test_chromosome[(courses_bits + classType_bits + sections_bits) : (courses_bits + classType_bits + sections_bits + max_students_bits)]
    test_chromosome_teacher = test_chromosome[(courses_bits + classType_bits + sections_bits+max_students_bits) : (courses_bits + classType_bits + sections_bits+max_students_bits + teachers_bits)]
    day1 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits)]
    timeSlot1 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits)]
    room1 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits)]
    room_size1 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)]
    day2 = test_chromosome[(courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits) : (courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits)]
    timeSlot2 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits)]
    room2 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits)]
    room_size2 = test_chromosome[(courses_bits + classType_bits + sections_bits  + max_students_bits + teachers_bits+ days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits) : (courses_bits + classType_bits + sections_bits  + max_students_bits+ teachers_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits + days_bits + timeSlots_bits + rooms_bits + roomSizeMax_bits)]
        
    
    #performing mutation
        
    # day1
    day1 = bin(randint(1, days))[2:]
    i = days_bits
    k=0
    j=len(day1)
    while i>0:
        i -= 1
        if j >= len(day1):
            if day1[k] == "1" and days_binary[k] == "1":
                day1 = day1[:k] + "1" + day1[k+1:]            
        else:
            day1 = day1[:k] + "0" + day1[k+1:]
    
        while i==0 and len(day1) < days_bits:  
            day1 = "0" + day1  
                
    if(int(test_chromosome_classType,2)==1):
        timeSlot1 = bin(randint(1, 4))[2:]
    else:
        timeSlot1 = bin(randint(3, timeSlots))[2:]
    i = timeSlots_bits
    k=0
    j=len(timeSlot1)
    while i>0:
        i -= 1
        if j >= len(timeSlot1):
            if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
        else:
            timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]
    
        while i==0 and len(timeSlot1) < timeSlots_bits:  
            timeSlot1 = "0" + timeSlot1 
                
    # room1
    room1 = bin(randint(1, rooms))[2:]
    i = rooms_bits
    k=0
    j=len(room1)
    while i>0:
        i -= 1
        if j >= len(room1):
            if room1[k] == "1" and rooms_binary[k] == "1":
                room1 = room1[:k] + "1" + room1[k+1:]            
        else:
            room1 = room1[:k] + "0" + room1[k+1:]
    
        while i==0 and len(room1) < rooms_bits:  
            room1 = "0" + room1     
            
    
    if room1 in halls:
        room_size1 = 120
    else:
        room_size1 = 60
    
   
   
     # day2
    day2 = bin(randint(1, days))[2:]
    i = days_bits
    k=0
    j=len(day2)
    while i>0:
        i -= 1
        if j >= len(day2):
            if day2[k] == "1" and days_binary[k] == "1":
                day2 = day2[:k] + "1" + day2[k+1:]            
        else:
            day2 = day2[:k] + "0" + day2[k+1:]
    
        while i==0 and len(day2) < days_bits:  
            day2 = "0" + day2 
            
    # timeSlot2
    
    if(int(test_chromosome_classType,2)==1):
        timeSlot2 = bin(randint(1, 4))[2:]
    else:
        timeSlot2 = bin(randint(3, timeSlots))[2:]
    i = timeSlots_bits
    k=0
    j=len(timeSlot2)
    while i>0:
        i -= 1
        if j >= len(timeSlot2):
            if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
        else:
            timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]
    
        while i==0 and len(timeSlot2) < timeSlots_bits:  
            timeSlot2 = "0" + timeSlot2 
                
    # room2
    room2 = bin(randint(1, rooms))[2:]
    i = rooms_bits
    k=0
    j=len(room2)
    while i>0:
        i -= 1
        if j >= len(room2):
            if room2[k] == "1" and rooms_binary[k] == "1":
                room2 = room2[:k] + "1" + room2[k+1:]            
        else:
            room2 = room2[:k] + "0" + room2[k+1:]
    
        while i==0 and len(room2) < rooms_bits:  
            room2 = "0" + room2  
    
    if room2 in halls:
        room_size2 = 120
    else:
        room_size2 = 60
        
    # room_size1
    room_size1 = bin(room_size1)[2:]
    i = roomSizeMax_bits
    k=0
    j=len(room_size1)
    while i>0:
        i -= 1
        if j >= len(room_size1):
            if room_size1[k] == "1" and roomSizeMax_binary[k] == "1":
                room_size1 = room_size1[:k] + "1" + room_size1[k+1:]            
        else:
            room_size1 = room_size1[:k] + "0" + room_size1[k+1:]
    
        while i==0 and len(room_size1) < roomSizeMax_bits:  
            room_size1 = "0" + room_size1
            
            
    # room_size2
    room_size2 = bin(room_size2)[2:]
    i = roomSizeMax_bits
    k=0
    j=len(room_size2)
    while i>0:
        i -= 1
        if j >= len(room_size2):
            if room_size2[k] == "1" and roomSizeMax_binary[k] == "1":
                room_size2 = room_size1[:k] + "1" + room_size2[k+1:]            
        else:
            room_size2 = room_size2[:k] + "0" + room_size2[k+1:]
    
        while i==0 and len(room_size2) < roomSizeMax_bits:  
            room_size2 = "0" + room_size2 
    
    mutated_chromosome = test_chromosome_course + test_chromosome_classType + test_chromosome_section + test_chromosome_class_strength + test_chromosome_teacher + day1 + timeSlot1 + room1 + room_size1 + day2 + timeSlot2 + room2 + room_size2
    if test_chromosome_classType == '1':
        if int(timeSlot1, 2) == 6:
            
            # timeSlot1
            timeSlot1 = bin(5)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot1)
            while i>0:
                i -= 1
                if j >= len(timeSlot1):
                    if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
                else:
                    timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]

                while i==0 and len(timeSlot1) < timeSlots_bits:  
                    timeSlot1 = "0" + timeSlot1
            # timeSlot2
            timeSlot2 = bin(int(timeSlot1, 2) + 1)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2        
            
            mutated_chromosome = test_chromosome_course + test_chromosome_classType + test_chromosome_section + test_chromosome_class_strength + test_chromosome_teacher + day1 + timeSlot1 + room1 + room_size1 + day2 + timeSlot2 + room1 + room_size1

        else:
            # timeSlot2
            timeSlot2 = bin(int(timeSlot1, 2) + 1)[2:]
            i = timeSlots_bits
            k=0
            j=len(timeSlot2)
            while i>0:
                i -= 1
                if j >= len(timeSlot2):
                    if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
                else:
                    timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

                while i==0 and len(timeSlot2) < timeSlots_bits:  
                    timeSlot2 = "0" + timeSlot2
            mutated_chromosome = test_chromosome_course + test_chromosome_classType + test_chromosome_section + test_chromosome_class_strength + test_chromosome_teacher + day1 + timeSlot1 + room1 + room_size1 + day2 + timeSlot2 + room1 + room_size1

     # checking boundary values
    # for time slots
    if int(timeSlot1, 2) > timeSlots:
    # timeSlot1
        timeSlot1 = bin(5)[2:]
        i = timeSlots_bits
        k=0
        j=len(timeSlot1)
        while i>0:
            i -= 1
            if j >= len(timeSlot1):
                if timeSlot1[k] == "1" and timeSlots_binary[k] == "1":
                        timeSlot1 = timeSlot1[:k] + "1" + timeSlot1[k+1:]            
            else:
                timeSlot1 = timeSlot1[:k] + "0" + timeSlot1[k+1:]

            while i==0 and len(timeSlot1) < timeSlots_bits:  
                timeSlot1 = "0" + timeSlot1
        
    if int(timeSlot2, 2) > timeSlots:
    # timeSlot2
        timeSlot2 = bin(5)[2:]
        i = timeSlots_bits
        k=0
        j=len(timeSlot2)
        while i>0:
            i -= 1
            if j >= len(timeSlot2):
                if timeSlot2[k] == "1" and timeSlots_binary[k] == "1":
                    timeSlot2 = timeSlot2[:k] + "1" + timeSlot2[k+1:]            
            else:
                timeSlot2 = timeSlot2[:k] + "0" + timeSlot2[k+1:]

            while i==0 and len(timeSlot2) < timeSlots_bits:  
                timeSlot2 = "0" + timeSlot2
    # for rooms          
    if int(room1, 2) > rooms :
        # room1
        room1 = bin(rooms)[2:]
        i = rooms_bits
        k=0
        j=len(room1)
        while i>0:
            i -= 1
            if j >= len(room1):
                if room1[k] == "1" and rooms_binary[k] == "1":
                    room1 = room1[:k] + "1" + room1[k+1:]            
            else:
                room1 = room1[:k] + "0" + room1[k+1:]

            while i==0 and len(room1) < rooms_bits:  
                room1 = "0" + room1     
            
    if int(room2, 2) > rooms :
        # room2
        room2 = bin(rooms)[2:]
        i = rooms_bits
        k=0
        j=len(room2)
        while i>0:
            i -= 1
            if j >= len(room2):
                if room2[k] == "1" and rooms_binary[k] == "1":
                    room2 = room2[:k] + "1" + room2[k+1:]            
            else:
                room2 = room2[:k] + "0" + room2[k+1:]

            while i==0 and len(room2) < rooms_bits:  
                room2 = "0" + room2  
    # for days
    if int(day2, 2) > days :
        # day2
        day2 = bin(days)[2:]
        i = days_bits
        k=0
        j=len(day2)
        while i>0:
            i -= 1
            if j >= len(day2):
                if day2[k] == "1" and days_binary[k] == "1":
                    day2 = day2[:k] + "1" + day2[k+1:]            
            else:
                day2 = day2[:k] + "0" + day2[k+1:]

            while i==0 and len(day2) < days_bits:  
                day2 = "0" + day2 
                
    if int(day1, 2) > days :
        # day1
        day1 = bin(days)[2:]
        i = days_bits
        k=0
        j=len(day1)
        while i>0:
            i -= 1
            if j >= len(day1):
                if day1[k] == "1" and days_binary[k] == "1":
                    day1 = day1[:k] + "1" + day1[k+1:]            
            else:
                day1 = day1[:k] + "0" + day1[k+1:]

            while i==0 and len(day1) < days_bits:  
                day1 = "0" + day1 
    
    if room2 in halls:
        room_size2 = 120
    else:
        room_size2 = 60
    
    if room1 in halls:
        room_size1 = 120
    else:
        room_size1 = 60
    
    return mutated_chromosome

def Genetic_Reproduction():
    
    Initialize_population()

    # Initialize a counter for consecutive iterations without fitness update
    consecutive_no_update = 0
    previous_fitness=[]


    for i in range(pop):
        fitness[i + 1] = calculateFitness(chromosome[i], chromosome)

        # Print fitness for debugging
    for i, fit in fitness.items():
        print(f"Chromosome {i}: Fitness {fit}")

    print("\nloopstart\n")
    # Loop until termination conditions are met
    while True:
        # Update fitness for all chromosomes
        for i in range(pop):
            fitness[i + 1] = calculateFitness(chromosome[i], chromosome)


        # Check for termination conditions
        if all(fit == 0 for fit in fitness.values()):
            print("All chromosomes have reached fitness 0. Termination.")
            break

        # Check for consecutive iterations without fitness update
        if previous_fitness==list(fitness.values()):
            consecutive_no_update += 1
            if consecutive_no_update >= 50:
                print("No update in fitness for 50 consecutive iterations. Termination.")
                break
        else:
            previous_fitness=list(fitness.values())
            consecutive_no_update = 0

        max_previous_fitness = max(fitness.values())

        # Perform tournament selection, crossover, and mutation
        tournament_size=5

        print("A")
        

        worst_chromosomes = tournament_selection(fitness, tournament_size)
        chromosome1, chromosome2 = worst_chromosomes

        child1, child2 = perform_crossover(chromosome[chromosome1 - 1], chromosome[chromosome2 - 1])


        child1_fitness = calculateFitness(child1, chromosome)
        child2_fitness = calculateFitness(child2, chromosome)

        if child1_fitness > fitness[chromosome1]:
            chromosome[chromosome1 - 1] = child1
            fitness[chromosome1] = child1_fitness
        #else:
        mutated_child1 = perform_mutation(child1)

        mutated_child1_fitness = calculateFitness(mutated_child1, chromosome)
        if mutated_child1_fitness > fitness[chromosome1]:
            chromosome[chromosome1 - 1] = mutated_child1
            fitness[chromosome1] = mutated_child1_fitness

        if child2_fitness > fitness[chromosome2]:
            chromosome[chromosome2 - 1] = child2
            fitness[chromosome2] = child2_fitness
        #else:
        mutated_child2 = perform_mutation(child2)

        mutated_child2_fitness = calculateFitness(mutated_child2, chromosome)
        if mutated_child2_fitness > fitness[chromosome2]:
            chromosome[chromosome2 - 1] = mutated_child2
            fitness[chromosome2] = mutated_child2_fitness

    print("\nfinal\n")
    for i, fit in fitness.items():
            print(f"Chromosome {i}: Fitness {fit}")
    num_rooms=rooms

def display_timetable(chromosome):
    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
    time_slots = ['08:30 AM - 09:50 AM', '10:00 AM - 11:20 AM', '11:30 AM - 12:50 PM', '01:00 PM - 02:20 PM', '02:30 PM - 03:50 PM', '04:00 PM - 05:20 PM']

    # Calculate the maximum width of the time slots and the "Free" text
    max_time_slot_width = max(len(slot) for slot in time_slots)
    free_text_width = len("Free")

    # Printing the timetable for each day
    for day in days:
        print(f"\n{day}:")  # Print the day
        # Printing the header row with time slots
        print("Room   ", end='')  # Space for the first cell
        for time_slot in time_slots:
            print(f"{time_slot:<{max_time_slot_width}}", end='')  # Fixed width for time slots
        print()

        for room in range(1, num_rooms + 1):  # Loop through each room
            print(f"Room {room:<2}", end='')  # Fixed width of 2 characters for room number
            for time_slot in range(6):  # Loop through each time slot
                course_info = None
                for chromosome_info in chromosome:
                    day_index = int(chromosome_info[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits):(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits)], 2)
                    time_slot_index = int(chromosome_info[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits):(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits)], 2)
                    room_index = int(chromosome_info[(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits):(courses_bits + classType_bits + sections_bits + teachers_bits + max_students_bits + days_bits + timeSlots_bits + rooms_bits)], 2)

                    if day_index == time_slot and room_index == room:
                        course_info = chromosome_info
                        break

                if course_info:
                    course = int(course_info[:courses_bits], 2)
                    class_type = int(course_info[courses_bits:(courses_bits + classType_bits)], 2)
                    section = int(course_info[(courses_bits + classType_bits):(courses_bits + classType_bits + sections_bits)], 2)
                    teacher = int(course_info[(courses_bits + classType_bits + sections_bits + max_students_bits):(courses_bits + classType_bits + sections_bits + max_students_bits + teachers_bits)], 2)
                    print(f"{course}-{class_type}-{section}-{teacher:<{free_text_width}}", end='')  # Fixed width for course info
                    print(" " * (max_time_slot_width - len(f"{course}-{class_type}-{section}-{teacher:<{free_text_width}}")), end='')  # Space for proper alignment
                else:
                    print("Free" + " " * (max_time_slot_width - free_text_width), end='')  # Fixed width for "Free" text
            print()



# Assuming you have 'chromosome' list, 'num_rooms' variable, and other constants defined
Genetic_Reproduction()
display_timetable(chromosome)


```
